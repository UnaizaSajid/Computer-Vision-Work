```python
import cv2
import numpy as np
from keras.preprocessing.image import img_to_array
from keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt

img = cv2.imread(r'C:\Users\Home\-------',0)
my=cv2.imshow('image',img)

plt.imshow(img, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```

```python
#--------------ROTATE IMAGE BY 180------------------

#image = cv2.rotate(img, cv2.cv2.ROTATE_180) 
image=cv2.rotate(img, cv2.ROTATE_90_COUNTERCLOCKWISE)  #rotate 270 degree
  
cv2.imwrite('13.png',image)
# Displaying the image 
plt.imshow(image, cmap = 'gray', interpolation = 'bicubic') 
plt.show()
```

```python
#---------------CROPPING ------------
rows,cols = img.shape
pts1 = np.float32([[56,65],[368,52],[28,387],[389,390]])
pts2 = np.float32([[0,0],[10,0],[0,10],[10,10]])
M = cv2.getPerspectiveTransform(pts1,pts2)
dst = cv2.warpPerspective(img,M,(50,50))
plt.subplot(121),plt.imshow(img ,cmap = 'gray', interpolation = 'bicubic'),plt.title('Input')
plt.subplot(122),plt.imshow(dst, cmap = 'gray', interpolation = 'bicubic'),plt.title('Output')
plt.show()
```

```python
#--------------- Random Rotation by an angle --------------------
def rotate(image, angle, center = None, scale = 1.0):
    (h, w) = image.shape[:2]

    if center is None:
        center = (w / 2, h / 2)

    # Perform the rotation
    M = cv2.getRotationMatrix2D(center, angle, scale)
    rotated = cv2.warpAffine(image, M, (w, h))
    plt.subplot(121),plt.imshow(img ,cmap = 'gray', interpolation = 'bicubic'),plt.title('Input')
    plt.subplot(122),plt.imshow(rotated, cmap = 'gray', interpolation = 'bicubic'),plt.title('Output')
    plt.show()

rotate(img,45)
```
