```python
import cv2
import os
import matplotlib.pyplot as plt


inputdir = r'C:\Users\Home\--------'
path=inputdir+"\\"
print(inputdir)

outdir = r'C:\Users\Home\---------'
path = outdir+"\\"
for filename in os.listdir(inputdir):
    image = cv2.imread(os.path.join(inputdir , filename))
    
    print(image.shape)
    
    dim = (224,224)
    
    resized_img = cv2.resize(image, dim, interpolation = cv2.INTER_AREA)
    plt.imshow(resized_img, cmap=plt.cm.gray)
  
    print('Resized Dimensions : ',resized_img.shape) 
    plt.show()
    cv2.imwrite(os.path.join(outdir , filename) ,resized_img)
```
